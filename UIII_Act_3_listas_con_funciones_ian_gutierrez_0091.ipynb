```python
# MENU DE PILOTO
# usando funciones y while True

pilotos = []


# FUNCION AGREGAR
def agregarpiloto():

    while True:

        nombre = input("ingresa el nombre del piloto: ")
        aerolinea = input("ingresa la aerolinea: ")
        avion = input("ingresa el modelo del avion: ")
        numero_vuelo = input("ingresa el numero de vuelo: ")
        horas_vuelo = int(input("ingresa las horas de vuelo: "))

        piloto = {
            "nombre": nombre,
            "aerolinea": aerolinea,
            "avion": avion,
            "numero_vuelo": numero_vuelo,
            "horas_vuelo": horas_vuelo
        }

        pilotos.append(piloto)

        print("piloto agregado correctamente")

        respuesta = input("deseas agregar otro piloto? si/no: ")

        if respuesta == "no":
            break


# FUNCION LEER
def leerpiloto():

    while True:

        nombre = input("nombre del piloto: ")
        aerolinea = input("aerolinea: ")
        avion = input("modelo del avion: ")
        numero_vuelo = input("numero de vuelo: ")
        horas_vuelo = int(input("horas de vuelo: "))

        print("datos del piloto")
        print("nombre:", nombre)
        print("aerolinea:", aerolinea)
        print("avion:", avion)
        print("numero de vuelo:", numero_vuelo)
        print("horas de vuelo:", horas_vuelo)

        respuesta = input("leer otro piloto? si/no: ")

        if respuesta == "no":
            break


# FUNCION LISTAR
def listarpilotos():

    while True:

        if len(pilotos) == 0:

            print("no hay pilotos registrados")

        else:

            print("lista de pilotos")

            for i in range(len(pilotos)):

                print(i + 1, pilotos[i])

        respuesta = input("volver a listar? si/no: ")

        if respuesta == "no":
            break


# FUNCION BUSCAR
def buscarpiloto():

    while True:

        buscar = input("ingresa el nombre del piloto a buscar: ")

        encontrado = False

        for piloto in pilotos:

            if piloto["nombre"] == buscar:

                print("piloto encontrado")
                print(piloto)

                encontrado = True

        if encontrado == False:

            print("piloto no encontrado")

        respuesta = input("buscar otro piloto? si/no: ")

        if respuesta == "no":
            break


# FUNCION MODIFICAR
def modificarpiloto():

    while True:

        buscar = input("ingresa el nombre del piloto a modificar: ")

        encontrado = False

        for piloto in pilotos:

            if piloto["nombre"] == buscar:

                piloto["nombre"] = input("nuevo nombre: ")
                piloto["aerolinea"] = input("nueva aerolinea: ")
                piloto["avion"] = input("nuevo avion: ")
                piloto["numero_vuelo"] = input("nuevo numero de vuelo: ")
                piloto["horas_vuelo"] = int(input("nuevas horas de vuelo: "))

                print("piloto modificado")

                encontrado = True

        if encontrado == False:

            print("piloto no encontrado")

        respuesta = input("modificar otro piloto? si/no: ")

        if respuesta == "no":
            break


# FUNCION BORRAR
def borrarpiloto():

    while True:

        buscar = input("ingresa el nombre del piloto a borrar: ")

        encontrado = False

        for piloto in pilotos:

            if piloto["nombre"] == buscar:

                pilotos.remove(piloto)

                print("piloto eliminado")

                encontrado = True

                break

        if encontrado == False:

            print("piloto no encontrado")

        respuesta = input("borrar otro piloto? si/no: ")

        if respuesta == "no":
            break


# MENU PRINCIPAL
while True:

    print("\nMENU DE PILOTOS DE AVION")
    print("1. agregar piloto")
    print("2. leer piloto")
    print("3. listar pilotos")
    print("4. buscar piloto")
    print("5. modificar piloto")
    print("6. borrar piloto")
    print("7. salir")

    opcion = int(input("ingresa una opcion: "))

    if opcion == 1:
        agregarpiloto()

    elif opcion == 2:
        leerpiloto()

    elif opcion == 3:
        listarpilotos()

    elif opcion == 4:
        buscarpiloto()

    elif opcion == 5:
        modificarpiloto()

    elif opcion == 6:
        borrarpiloto()

    elif opcion == 7:
        print("fin del programa")
        break

    else:
        print("opcion no valida")

        print("Software de Control de piloto finalizado")
        print("Desarrollado por: Ian Gutiérrez | NL: 22")
```
